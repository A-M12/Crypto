```python
def cryptanalysis():
  cipher_text=input('Enter the cipher text for cryptanalysis : ')
  for k in range(26):
    plain_text=''
    for letter in cipher_text:
      if letter==' ':
        plain_text+=letter
      else:
        c=ord(letter)-65
        e=(c-k)%26
        plain_text+=chr(e+65)
    print('With key = ',k,plain_text)
cryptanalysis()
```
